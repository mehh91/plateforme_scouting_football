## Premières conclusions de l’audit events

- La table events est une table large et hétérogène.
- La colonne `type` est centrale pour segmenter les actions.
- Les colonnes spécifiques (`pass_*`, `shot_*`, `carry_*`, etc.) ne sont remplies que pour certains types d’événements.
- Les coordonnées terrain (`location`, `pass_end_location`, `carry_end_location`, `shot_end_location`) seront clés pour le feature engineering.
- Les types d’événements prioritaires pour la V1 seront probablement :
  - Pass
  - Shot
  - Carry
  - Interception
  - Ball Recovery
  - Duel
- Un travail de parsing et de sélection de colonnes sera nécessaire avant de construire les features.
